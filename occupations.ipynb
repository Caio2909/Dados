```python
import pandas as pd

users= pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|", index_col="user_id")
users
```

```python
userMeanAge = users.groupby("occupation").mean(True)
userMeanAge
```

```python
gender_ocup = users.groupby(['occupation','gender']).agg({'gender':'count'})
occup_count = users.groupby(['occupation']).agg('count')
occup_gener = gender_ocup.div(occup_count, level='occupation') * 100
occup_gener.loc[:, 'gender']
```

```python
maleRatio = users.groupby("occupation")
maleRatio = maleRatio['gender'].apply(lambda x: (x == "M").mean()*100).sort_values(ascending=False).round(1)
femaleRatio = 100 - maleRatio
maleRatio
ratios= {'M': maleRatio,'F': femaleRatio }
MFRatio = pd.DataFrame(ratios)
MFRatio
```

```python
minMax = users.groupby("occupation").age.agg(["min", "max"])
minMax
```

```python
userMeanAgeGender = users.groupby(["occupation","gender"]).age.mean(True)
userMeanAgeGender
```
